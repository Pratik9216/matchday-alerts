```
import pandas as pd 
```

```
df = pd.read_csv('/content/drive/MyDrive/Inframind/Customer engagement/original_dataset/processed_dataset.csv')
df.head()
```

```
df.drop(['match_id', 'inning', 'batting_team', 'batsman', 'bowler', 'player_dismissed',
        'batsman_runs','is_wicket','boundary_next_12_balls'], axis = 1, inplace = True) 
df.head()
```

```
# filter and clean bowling_average 
for i in df.index:
  if df['bowling_average'][i] == '-':
    df.at[i, 'bowling_average'] = 40.00
```

```
# filter and clean strike_rate 
for i in df.index:
  if df['strike_rate'][i] == '-':
    df.at[i, 'strike_rate'] = 60.00
```

```
# one hot encoding of col "bowling_team"

# Get one hot encoding of col "bowling_team "
one_hot = pd.get_dummies(df['bowling_team'], prefix='bowling_team')
# Drop that column as it is now encoded
df = df.drop('bowling_team',axis = 1)
# Join the encoded one with df
df = df.join(one_hot)
df.head()
```
